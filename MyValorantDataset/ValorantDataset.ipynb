```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import GridSearchCV
%matplotlib inline
```

```python
Valorant_data = pd.read_csv("Valorant5.csv")
```

```python
Valorant_data.columns.to_list()
```

```python
Valorant_data.describe()
```

```python
y = Valorant_data.Won
X = Valorant_data.drop(["Won","Serge_peak"],axis=1)
```

```python
parametrs = {"criterion":["entropy","gini"],"max_depth":range(1,3),"min_samples_split":range(2,3),"min_samples_leaf":range(1,5)}
```

```python
clf = DecisionTreeClassifier()
```

```python
search = GridSearchCV(clf,parametrs)
search.fit(X,y)
```

```python
best_tree = search.best_estimator_
```

```python
plot_tree(best_tree, feature_names=list(X),class_names=['Lose', 'Won'], filled=True)
plt.show()
```
